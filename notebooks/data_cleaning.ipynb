```python
# === Imports ===
import pandas as pd
import os
import sys

# Add src to path to import custom modules
sys.path.append('../src')

from data_loader import load_all_data
from preprocessing import (
    clean_fraud_data,
    clean_creditcard_data,
    clean_ip_country_data
)

# === Set display options ===
pd.set_option('display.max_columns', 100)

# === Define paths ===
RAW_DATA_PATH = '../data/raw'
PROCESSED_DATA_PATH = '../data/processed'
os.makedirs(PROCESSED_DATA_PATH, exist_ok=True)
```

```python
# === Load raw data ===
fraud_df_raw, credit_df_raw, ip_df_raw = load_all_data(RAW_DATA_PATH)
```

```python
# === Clean datasets using preprocessing module ===
fraud_df_cleaned = clean_fraud_data(fraud_df_raw)
credit_df_cleaned = clean_creditcard_data(credit_df_raw)
ip_df_cleaned = clean_ip_country_data(ip_df_raw)
```

```python
# === Preview cleaned data ===
print("\n Cleaned Fraud Data:")
display(fraud_df_cleaned.head())
print(fraud_df_cleaned.info())

print("\n Cleaned Credit Card Data:")
display(credit_df_cleaned.head())
print(credit_df_cleaned.info())

print("\n Cleaned IP-Country Data:")
display(ip_df_cleaned.head())
print(ip_df_cleaned.info())
```

```python
# === Save cleaned datasets ===
fraud_df_cleaned.to_csv(f"{PROCESSED_DATA_PATH}/fraud_data_cleaned.csv", index=False)
credit_df_cleaned.to_csv(f"{PROCESSED_DATA_PATH}/creditcard_cleaned.csv", index=False)
ip_df_cleaned.to_csv(f"{PROCESSED_DATA_PATH}/ip_country_merged.csv", index=False)

print("\n📁 Cleaned datasets saved to ../data/processed/")
```
